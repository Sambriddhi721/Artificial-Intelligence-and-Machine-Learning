Task- 1:<br>

```python
def convert_units(conversion_type, value):
    """
    Converts between different units of measurement.

    Parameters:
    conversion_type (str): Type of conversion chosen by user
    value (float): numeric value to convert

    Returns:
    float: converted value
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084

    elif conversion_type == "ft_to_m":
        return value / 3.28084

    elif conversion_type == "kg_to_lb":
        return value * 2.20462

    elif conversion_type == "lb_to_kg":
        return value / 2.20462

    elif conversion_type == "l_to_gal":
        return value * 0.264172

    elif conversion_type == "gal_to_l":
        return value / 0.264172

    else:
        raise ValueError("Unsupported conversion type")


def main():
    print("Choose conversion type:")
    print("1: meters → feet")
    print("2: feet → meters")
    print("3: kg → pounds")
    print("4: pounds → kg")
    print("5: liters → gallons")
    print("6: gallons → liters")

    choice = input("Enter choice (1-6): ")

    try:
        value = float(input("Enter value: "))

        conversions = {
            "1": "m_to_ft",
            "2": "ft_to_m",
            "3": "kg_to_lb",
            "4": "lb_to_kg",
            "5": "l_to_gal",
            "6": "gal_to_l"
        }

        result = convert_units(conversions[choice], value)

        print("Converted value:", result)

    except ValueError:
        print("Invalid input. Please enter numeric values.")

    except KeyError:
        print("Invalid conversion option.")


main()
```

```python
def find_sum(numbers):
    """Returns the sum of numbers"""
    return sum(numbers)


def find_average(numbers):
    """Returns the average of numbers"""
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """Returns the maximum number"""
    return max(numbers)


def find_min(numbers):
    """Returns the minimum number"""
    return min(numbers)


def main():

    try:
        nums = input("Enter numbers separated by space: ")

        numbers = [float(x) for x in nums.split()]

        if len(numbers) == 0:
            raise ValueError("List cannot be empty")

        print("Choose operation:")
        print("1 Sum")
        print("2 Average")
        print("3 Maximum")
        print("4 Minimum")

        choice = input("Enter choice: ")

        if choice == "1":
            print("Sum:", find_sum(numbers))

        elif choice == "2":
            print("Average:", find_average(numbers))

        elif choice == "3":
            print("Maximum:", find_max(numbers))

        elif choice == "4":
            print("Minimum:", find_min(numbers))

        else:
            print("Invalid choice")

    except ValueError:
        print("Invalid input. Enter numbers only.")


main()
```

4.3 Exercise on Nested List:
1. Flatten a Nested List:

```python
def flatten(lst):
    result = []
    for sub in lst:
        result.extend(sub)
    return result

print(flatten([[1,2],[3,4],[5]]))
```

2. Accessing Nested List Elements:


```python
def access_nested_element(lst, indices):
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

3. Sum of All Elements in a Nested List:

```python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item,list):
            total += sum_nested(item)
        else:
            total += item
    return total
print(sum_nested([[1,2],[3,[4,5]],6]))
```

4. Remove Specific Element from a Nested List:


```python
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item,list):
            result.append(remove_element(item,elem))
        else:
            if item != elem:
                result.append(item)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

5. Find the Maximum Element in a Nested List:


```python
def find_max(lst):
    maximum = float("-inf")

    for item in lst:
        if isinstance(item,list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum,item)

    return maximum

print(find_max([[1,2],[3,[4,5]],6]))
```

6. Count Occurrences of an Element in a Nested List:


```python
def count_occurrences(lst, elem):
    count = 0

    for item in lst:
        if isinstance(item,list):
            count += count_occurrences(item,elem)
        elif item == elem:
            count += 1

    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

7. Flatten a List of Lists of Lists:


```python
def deep_flatten(lst):
    result = []

    for item in lst:
        if isinstance(item,list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

8. Nested List Average:


```python
def average_nested(lst):

    flat = deep_flatten(lst)

    return sum(flat) / len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```

Problem- 4: Matrix Operations:<br>
• For the following arrays:<br>
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]) <br>

```python
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

print("A:\n", A)
print("B:\n", B)
```

1. Prove A.A−1 = I.


```python
A_inv = np.linalg.inv(A)
I = np.eye(A.shape[0])

print("A inverse:\n", A_inv)
print("A @ A_inv:\n", A @ A_inv)
print("Is A @ A_inv close to I?", np.allclose(A @ A_inv, I))
```

2. Prove AB̸ = BA.


```python
AB = A @ B
BA = B @ A

print("AB:\n", AB)
print("BA:\n", BA)
print("Is AB equal to BA?", np.array_equal(AB, BA))
```

3. Prove (AB)T = BTAT.


```python
left = (A @ B).T
right = B.T @ A.T

print("(AB)^T:\n", left)
print("B^T A^T:\n", right)
print("Are they equal?", np.allclose(left, right))
```
